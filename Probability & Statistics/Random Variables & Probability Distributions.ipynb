### Bernoulli Distribution

```python
import numpy as np
import matplotlib.pyplot as plt

p = 0.8
X = [0,1]
probs = [1 - p, p]

plt.bar(X, probs, color=['red', 'green'])
plt.xticks(X, ['Failure (0)', 'Success (1)'])
plt.xlabel("Outcome")
plt.ylabel("Probability")
plt.title("Bernoulli Distribution (p=0.7)")
plt.show()
```

```python
# Simulating Customer Purchase Behavior

customers = np.random.binomial(n = 1, p = 0.2, size = 1000)

# Plot
plt.bar(['No Purchase (0)', 'Purchase (1)'], [sum(customers == 0), sum(customers == 1)], color=['red', 'green'])
plt.xlabel("Customer Action")
plt.ylabel("Count")
plt.title("Bernoulli Distribution: Customer Purchase Behavior")
plt.show()
```

### Binomial Distribution

```python
from scipy.stats import binom
import seaborn as sns

n, p = 10, 0.5
x = np.arange(0, n+1)
y = binom(x, n, p)

sns.barplot(x=x, y=y)
plt.xlabel("Number of successes")
plt.ylabel("Probability")
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.show()
```

### Poisson Distribution

```python
from scipy.stats import poisson

lambda_ = 5  # Expected number of events per unit time
x = np.arange(0, 15)
y = poisson.pmf(x, lambda_)

sns.barplot(x=x, y=y)
plt.xlabel("Number of events")
plt.ylabel("Probability")
plt.title("Poisson Distribution (=5)")
plt.show()
```

### Normal Distribution

```python
from scipy.stats import norm

mu, sigma = 0, 1  # Mean and standard deviation
x = np.linspace(-4, 4, 100)
y = norm.pdf(x, mu, sigma)

plt.plot(x, y)
plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.title("Normal Distribution (=0, =1)")
plt.show()
```

### Exponential Distribution

```python
from scipy.stats import expon

lambda_ = 0.5  # Mean time until failure
x = np.linspace(0, 10, 1000)
y = expon.pdf(x, scale=1/lambda_)

plt.plot(x, y)
plt.xlabel("Time Until Website Fails")
plt.ylabel("Probability Density")
plt.title("Exponential Distribution: Website Downtime Prediction")
plt.show()
```

### Uniform Distribution

```python
import numpy as np

clicks = np.random.uniform(0, 100, 1000)  # Click rates between 0-100
plt.hist(clicks, bins=20, color="blue", edgecolor="black")
plt.xlabel("Click Rate")
plt.ylabel("Frequency")
plt.title("Uniform Distribution: Random User Clicks")
plt.show()
```
